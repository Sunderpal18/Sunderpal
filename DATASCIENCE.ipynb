### **Exploratory Data Analysis (EDA) Summary**  

I performed **Exploratory Data Analysis (EDA)** on the **Olympic Athletes Dataset**, which included **Univariate and Multivariate Analysis** to understand the data's structure, distribution, and relationships.  

---

## **1. Data Loading & Inspection**  
- Loaded the dataset and displayed basic information (df.info()).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("athletes.csv")
df.head()
```

## Checked for missing values 

```python
# Check for missing values
print("Missing values:")
print(df.isnull().sum())
```

### getting the info of the data

```python
# Display basic info
df.info()
```

### **Findings:** 
- The dataset contains **11,113 rows** and **36 columns**.  
- Missing values exist in some columns, especially in height and weight.  
- The key numerical variables are **height** and **weight**, while categorical variables include **gender** and **country**. 


## **2. Univariate Analysis (Single Variable Analysis)**  

### **A. Distribution of Height**  
- Used a **histogram with KDE** to visualize the height distribution.  
 **Findings:**  
- Most athletes have a height concentrated in a specific range.  
- There are some outliers, indicating very tall or very short athletes.  



```python
# Univariate Analysis
plt.figure(figsize=(8,5))
sns.histplot(df['height'].dropna(), bins=30, kde=True)
plt.title("Distribution of Athletes' Height")
plt.show()
```

### **B. Distribution of Weight**  
- Similar analysis as height but focused on weight.  
 **Findings:**  
- Weight follows a normal distribution but has some extreme values.  

```python
plt.figure(figsize=(8,5))
sns.histplot(df['weight'].dropna(), bins=30, kde=True)
plt.title("Distribution of Athletes' Weight")
plt.show()
```

### **C. Gender Distribution**  
- Used a **count plot** to visualize the distribution of male and female athletes.  
 **Findings:**  
- The dataset has a slightly **higher number of male athletes** than female athletes.  

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df['gender'])
plt.title("Gender Distribution")
plt.show()
```

## **3. Multivariate Analysis (Relationship Between Variables)**  

### **A. Height vs. Weight (Scatter Plot by Gender)**  
- Created a **scatter plot** to analyze the relationship between height and weight.  
**Findings:**  
- There is a **positive correlation** between height and weight (taller athletes tend to be heavier).  
- Males generally have **higher height and weight** compared to females.

```python
# Multivariate Analysis
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['height'], y=df['weight'], hue=df['gender'])
plt.title("Height vs Weight by Gender")
plt.show()
```

### **B. Top 10 Countries with the Most Athletes (Bar Chart)**  
- Analyzed which countries have the highest representation in the dataset.  
**Findings:**  
- A few countries dominate the dataset, with the **USA, China, and Russia** having the most athletes. 

```python
plt.figure(figsize=(12,6))
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.xticks(rotation=45)
plt.title("Top 10 Countries with Most Athletes")
plt.show()
```

### **C. Correlation Heatmap (Numerical Variables)**  
- Used a **heatmap** to see correlations between numerical variables.  
**Findings:**  
- Height and weight show a **strong positive correlation** (as expected).  
- Other variables might have weak or no correlation with height and weight. 

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[['height', 'weight']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Height and Weight")
plt.show()
```

## **Key Insights from the Analysis**  
1. **Height and weight** have a strong correlation, which is expected in sports.  
2. **Gender distribution** shows slightly more male athletes than female athletes.  
3. **Certain countries** dominate the dataset, suggesting a higher participation rate in the Olympics.  
4. **Some extreme values (outliers)** exist, indicating the need for further investigation or cleaning.  

This analysis helps us **understand athlete characteristics and country-wise participation trends**
